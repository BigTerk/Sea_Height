```python
# Read in the Sheets 

# test it
#df1 = pd.read_excel(xls, 'Sheet1')
#df2 = pd.read_excel(xls, 'Sheet2')


# Real way 
xls = pd.ExcelFile('C:\\Users\\gst8882\\Desktop\\NOAA_Test.xlsx')
# make a list with each sheet as an item in the list
#df_list = []
df_list=pd.DataFrame()
# loop through the sheets
for i in range(1,37):
    #Sheet numbers
    s = 'Sheet'+str(i)    
    #Build the list 
    #df_list.append(pd.read_excel(xls, s))
    Raw = pd.read_excel(xls, s)
    
    Raw["City"]=[City[i-1]]*Raw.shape[0]
    Raw["State"]=[State[i-1]]*Raw.shape[0]
    Raw["Station"]=[Station[i-1]]*Raw.shape[0]
    
    if i ==1:
        df_list= Raw
    else:
        df_list=pd.concat([df_list, Raw])
```
